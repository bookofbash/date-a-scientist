```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Examine the data 
print(data['body_type'].value_counts())
print(data['orientation'].value_counts())
```

```python
# Distribition of body types
data['body_type'].value_counts().plot(kind='bar')
plt.title('Distribution of Body Types')
plt.xlabel('Body Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for orientation
data['orientation'].value_counts().plot(kind='bar', color='purple')
plt.title('Distribution of Sexual Orientations')
plt.xlabel('Orientation')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for Diet
data['diet'].value_counts().plot(kind='bar', color='purple')
plt.title('Distribution Diet')
plt.xlabel('Diet')
plt.ylabel('Frequency')
plt.show()
```

```python
#Cross Analysis
import seaborn as sns
cross_tab = pd.crosstab(data['body_type'], data['orientation'])
print(cross_tab)

# Heatmap of the cross-tabulation
plt.figure(figsize=(10, 8))
sns.heatmap(cross_tab, annot=True, cmap='coolwarm', fmt='d')
plt.title('Heatmap of Body Type vs. Orientation')
plt.show()
```

```python
#Looking at bisexual body type
bisexual_data = data[data['orientation'] == 'bisexual']

body_type_counts = bisexual_data['body_type'].value_counts()

#Plotting the bar chart
plt.figure(figsize=(10, 6))
body_type_counts.plot(kind='bar', color='purple')
plt.title('Distribution of Body Types Among Bisexual Individuals')
plt.xlabel('Body Type')
plt.ylabel('Number of Individuals')
plt.xticks(rotation=45)
plt.show()
```
